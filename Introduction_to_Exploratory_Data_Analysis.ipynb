```python
df=df[['coaster_name',
       #'Length', 'Speed',
       'Location', 'Status', #'Opening date',
       #'Type',
       'Manufacturer',
       #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date',
       #'Opened',
       #'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', #'height_value', 'height_unit',
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
df
```

```python
#converting to datetime
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
#Rename columns
# df.columns
df=df.rename(columns={
   'coaster_name':'Coaster_Name',
   'year_introduced':'Year_Introduced',
   'latitude':'Latitude',
   'longitude':'Longitude',
   'opening_date_clean':'Opening_Date',
   'speed_mph':'Speed_mph',
   'height_ft':'Height_ft',
   'Inversions_clean':'Inversions_Clean',
   'Gforce_clean':'Gforce_Clean'

}).copy()
df
```

```python
# checking for missing values/null
df.isna().sum()
```

```python
# checking for duplicated row values in data
# below code shows all columns that have duplicated values
df.loc[df.duplicated()]
```

```python
# trying to validate the columns given above in previous cell have the duplicated as indicated. so we are gonna take Coaster_Name	 and check if it has duplicates
# if we run the below command by adding df.loc it shows all the duplicated rows
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
#trying to validate using query that 	Crystal Beach Cyclone from Coaster_Name is duplicated. so the 1st value that appears is the main value and the 2nd is the duplicated
df.query('Coaster_Name=="Crystal Beach Cyclone"')
```

```python
# Remove any number of duplicates with certain number of columns that are the same
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
#now if Coaster_Name,	Location and Opening_Date are duplicated (here weve just picked an example of columns duplicated).we find the inverse(this means we just take only columns that are not duplicated in thats just the 1st rows values that had initial values before the duplicate)
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
df
```

# Step3 FEATURE UNDERSTANDING
(univariant analysis)
*   Histogram
*   KDE
*   Box plot





```python
#counts how many unique/duplicates values occur.automatically orders from most to least number of roller coaster introduced. shows what years had the most and least rollercoaster introduced
ax=df['Year_Introduced'].value_counts() \
.head(10) \
.plot(kind='bar' ,title='Top 10 Years Roller Coaster Introduced')
plt.xlabel('Year Introduced')
# ax.set_ylabel('frequency')
plt.ylabel('Count')
```

```python
ax=df['Speed_mph'].plot(kind='hist',bins=30 ,title='Roller Coaster speed(mph)')
plt.xlabel('Speed(mph)')
# ax.set_ylabel('frequency')
plt.ylabel('frequency')
```

```python
# using Kernel Densly plot(KDE)
ax=df['Speed_mph'].plot(kind='kde',title='Roller Coaster speed(mph)')
plt.xlabel('Speed(mph)')
# ax.set_ylabel('frequency')
plt.ylabel('frequency')
```

# **STEP 5: ASK QUESTION ABOUT THE DATA**

1.What are the locations with the fastest roller costers with min of 10 coasters at that place

```python
# locations=df['Location'].value_counts()
# (locations)
df.query('Location !="Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count>=10').sort_values('mean').plot(kind='barh',figsize=(12,5),title='Average Coaster Speed by Location')
plt.xlabel('Average Roller Coaster Speed')
plt.ylabel('Location')
plt.show()
```
